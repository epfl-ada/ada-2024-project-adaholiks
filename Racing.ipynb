```python
import pandas as pd
from pyarrow import feather
```

```python
paths = feather.read_feather('Data/dataframes/paths.feather')
```

```python
finished_paths = paths[paths['finished']]

# Initialize a dictionary to store the DataFrames
dfs_by_distance = {}

# Iterate over the range of distances
for d in range(1, int(finished_paths['distance'].max()) + 1):
    # Filter and sort the DataFrame for the current distance
    df_d = finished_paths[finished_paths['distance'] == d].sort_values(by='durationInSec')
    
    # Store the DataFrame in the dictionary with the key `d`
    dfs_by_distance[f"df_{d}"] = df_d

# Display the first few rows of the DataFrame for distance d
d = 6
top = dfs_by_distance[f"df_{d}"].head(20)
print(top['hashedIpAddress'].unique().shape)
print(top['identifier'].unique().shape)
top.head(20)

# NOTE: probably two friends challenged themselves to get this one path as fast as possible
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the data for the relevant hashed IP addresses
filtered_data = finished_paths[
    (finished_paths['identifier'] == 5634) &
    (finished_paths['hashedIpAddress'].isin(['62d5c0be1ee5c287', '3c4bf61c4a447176']))
].copy()

# Add an attempt number column (per hashedIpAddress, ranked by timestamp)
filtered_data['attempt'] = filtered_data.groupby('hashedIpAddress').cumcount() + 1

# Create the bar plot
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(
    data=filtered_data,
    x='attempt',
    y='durationInSec',
    hue='hashedIpAddress',
    ax=ax
)

# Add plot labels and a title
ax.set_title('Duration in Seconds per Attempt', fontsize=14)
ax.set_xlabel('Attempt Number', fontsize=12)
ax.set_ylabel('Duration (Seconds)', fontsize=12)
ax.legend(title='Hashed IP Address')

plt.show()
```

```python
import plotly.express as px

# Filter the data for the relevant hashed IP addresses
filtered_data = finished_paths[
    (finished_paths['identifier'] == 5634) &
    (finished_paths['hashedIpAddress'].isin(['62d5c0be1ee5c287', '3c4bf61c4a447176']))
].copy()

# Map hashed IP addresses to Player 1 and Player 2
ip_to_player = {
    '62d5c0be1ee5c287': 'Player 2',
    '3c4bf61c4a447176': 'Player 1'
}
filtered_data['Player'] = filtered_data['hashedIpAddress'].map(ip_to_player)

# Add an attempt number column (per player, ranked by timestamp)
filtered_data['attempt'] = filtered_data.groupby('Player').cumcount() + 1

# Create the interactive line plot
fig = px.line(
    filtered_data,
    x='attempt',
    y='durationInSec',
    color='Player',  # Use the new 'Player' column for legend
    title='Duration in Seconds per Attempt (Race Between Players)',
    labels={'Player': 'Player', 'attempt': 'Attempt Number', 'durationInSec': 'Time (Seconds)'}
)

# Customize the plot
fig.update_layout(
    xaxis_title='Attempt Number',
    yaxis_title='Time (Seconds)',
    legend_title_text='',
    hovermode='x unified'  # Allows hovering across data points to view details
)

# Show the plot
fig.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_attempt_durations(data, hashed_ip, identifier=None):
    """
    Plots attempt durations for a single hashed IP address, optionally filtering by identifier.
    Uses different colors for bars if multiple identifiers are present.
    
    Parameters:
        data (pd.DataFrame): The dataset containing 'hashedIpAddress', 'timestamp', 'durationInSec', and 'identifier'.
        hashed_ip (str): The hashed IP address to filter and plot.
        identifier (int, optional): The identifier to filter by. Default is None (no filtering by identifier).
    
    Returns:
        None: Displays the bar plot.
    """
    # Filter the data for the relevant hashed IP address
    filtered_data = data[data['hashedIpAddress'] == hashed_ip]
    
    # Apply identifier filtering if provided
    if identifier is not None:
        filtered_data = filtered_data[filtered_data['identifier'] == identifier]
    
    # Ensure there is data to plot
    if filtered_data.empty:
        print(f"No data found for Hashed IP: {hashed_ip} with Identifier: {identifier}")
        return
    
    # Add an attempt number column (ranked by timestamp)
    filtered_data = filtered_data.sort_values(by='timestamp')  # Ensure attempts are ordered by time
    filtered_data['attempt'] = range(1, len(filtered_data) + 1)
    
    # Create the bar plot
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.barplot(
        data=filtered_data,
        x='attempt',
        y='durationInSec',
        hue='identifier' if identifier is None else None,  # Use identifier for color only if not filtering
        palette='Set2',
        ax=ax
    )
    
    # Add plot labels and a title
    title = f'Duration in Seconds per Attempt for Hashed IP: {hashed_ip}'
    if identifier is not None:
        title += f' (Identifier: {identifier})'
    ax.set_title(title, fontsize=14)
    ax.set_xlabel('Attempt Number', fontsize=12)
    ax.set_ylabel('Duration (Seconds)', fontsize=12)
    
    # Adjust legend
    legend = ax.get_legend()
    if legend is not None and identifier is not None:
        legend.remove()  # Remove legend if only one identifier
    
    plt.show()

# Example usage:
plot_attempt_durations(finished_paths, '62d5c0be1ee5c287')  # Without identifier
#plot_attempt_durations(finished_paths, '62d5c0be1ee5c287', identifier=5634)  # With identifier
```
